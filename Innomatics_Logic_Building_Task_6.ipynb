```python
#Problem statement No:1
def Parking_system(logs, capacity):
    peak_use = 0
    current_parked = 0
    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
    
        if current_parked > peak_use:
            peak_use = current_parked
    
    if current_parked > capacity:
        print("The parking Capacity Exceeded!")
    else:
        print("Parking Space Available!")

    print(f"Current ocuppied Parking: {current_parked}")
    print(f"Peak usage : {peak_use}")

parking_logs = ["IN","IN","IN","IN","IN","IN","OUT","OUT","OUT","IN"]
capacity = int(input("Enter the capacity of the Parking lot:"))
Parking_system(parking_logs, capacity)
```

```python
#Problem No: 2
def delivery_time(distance, traffic, weather):
    # Let speed of vehicle be  = 40 km/hr
    base_time = (distance / 40) * 60  
    
    delay = 0
    if traffic == "Low":
        delay += 5
    elif traffic == "Medium":
        delay += 10
    elif traffic == "High":
        delay += 20

    if weather == "Rainy":
        delay += 10
    elif weather == "Storm":
        delay += 20

    eta = base_time + delay

    print(f"Estimated Delivery Time: {int(eta)} minutes")


# Sample Input
distance = 10
traffic = "Medium"
weather = "Rainy"

delivery_time(distance, traffic, weather)
```

```python
#Problem N0: 03
def movie_seat_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    
    occupancy = (booked_count / total_seats) * 100
    
    print(f"Occupancy: {int(occupancy)}%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull ")
    elif occupancy >= 75:
        print("Show Status: Almost Full🔥")
    elif occupancy >= 40:
        print("Show Status: Fast Filling")
    else:
        print("Show Status: Seats Available")

    if occupancy >= 90:
        print("Suggestion: Opening Additional Show")


total_seats = 300
booked_seats = [1] * 150   # 150 bookings are done

movie_seat_analyzer(total_seats, booked_seats)
```

```python
#problem Statement: 4
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading
    
    average = total / len(cpu_readings)
    
    print(f"Average Load: {int(average)}%")
    
    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


cpu_readings = [52, 65, 73, 89, 92]

classify_server_load(cpu_readings)
```

```python
#problem No: 5
def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if resource == "Projector" and hours > 5:
             overused.append(resource)

        elif resource == "AC" and hours > 8:
            overused.append(resource)

        elif resource == "Lights" and hours > 6:
            overused.append(resource)

    if overused: #if True
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused: None")
        print("Energy Alert: No ")


usage = {
    "Projector": 5,
    "AC": 10,
    "Lights":3
}

classroom_resource_monitor(usage)
```

```python
#Problem NO: 6

def event_registration(capacity, registrations):
    if registrations <= capacity:
        print("Confirm Registrations:", registrations)
        print("waitlist Users: 0")
        print("Registration Status: Open")
    else:
        Confirm = capacity
        waitlist = registrations - capacity

        print("Confirm Registrations:", Confirm)
        print("waitlist Users:", waitlist)
        print("Registration Status: Closed")


event_capacity = 250
registrations = 175

event_registration(event_capacity, registrations)
```
