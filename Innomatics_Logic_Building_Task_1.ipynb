1. User Login Check

```python
username = "admin"
password = "1234"

input_username = input("Enter username: ")

if input_username == username:
    print("have acess to this user")
else:
    print("doesn't have an acess to this user")

input_password = input("Enter password: ")
if input_password == password:
    print("password is correct and Login Successful")
else:
    print("Invalid password")
```

2. pass and fail analyzer

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

3. Simple data cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names = []
for name in names:
    name = name.strip() #cut off the extra space
    name = name.lower()
    cleaned_names.append(name)
print(cleaned_names)
```

4.Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)

    if length > 10:
        print("Long message")
        print()#to give an space between an output
    else:
        print("Short message")
        print()#to give an space between an output
```
