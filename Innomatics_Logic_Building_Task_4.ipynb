```python
"""1. Mobile Recharge Validation System:
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid"""
#functionfor mobile recharge
def mobile_recharge(amount,valid_plans):
  if amount < 50:
    return "Recharge amount must be at least ₹50."

  if amount not in valid_plans:
    print("Available plans:", valid_plans)
    return "Invalid plan selected"
  return "Recharge successful"
#list
valid_plans=[199,299,399,599]
#Logic
while True:
  amount=int(input("Enter amount:"))
  result=mobile_recharge(amount,valid_plans)
  print(result)
#if reacharge is invalid,allowing the user to retry
  if result=="Recharge successful":
    break
  else:
    print("Try again!")
```

```python
"""2.Inventory Reorder Alert System:
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise"""
#function
def Reorder_alert(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

stock={"A":20,"B":25,"C":30,"D":14}
Reorder_alert(stock)
```

```python
"""3.Student Result Processing System:
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average ≥ 50 else Fail"""
#function to calculate average marks
def total_marks(marks):
  total=0
  for mark in marks:
    total=total+mark
  avg=total/len(marks)
  print("Average marks:",avg)
  #logic to see whether a student is pass or fail
  if avg>=50:
    return "pass"
  else:
    return "fail"
marks=[50,70,40,80,90,99]
result=total_marks(marks)
print(result)
```

```python
"""4.Cab Fare Estimator with Retry Option:
Create a cab fare calculator:
Base fare = ₹50
₹12 per km
Peak hour → 25% extra
Allow retry using while loop"""
#function to calculate fare
def fare_calculation(distance,peak):
  base_fare=50
  km=12
  fare=base_fare+(distance*km)
  if peak.lower()=="yes":
    fare=fare+(fare*0.25)
  return fare
#while loop logic
while True:
  try:
    distance=float(input("Enter distance:"))
    peak=input("Is it peak hour(yes/no):")
    total_fare=fare_calculation(distance, peak)
    print("Total fare:",total_fare)
    again = input("Do you want to calculate again? (yes/no): ")
    if again.lower() != "yes":
      break
  except ValueError:
    print("Please enter valid numeric distance!")
    print("Try again...\n")
```

```python
"""5.Employee Attendance Eligibility Checker:
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if ≥ 75% else Not Eligible"""
#functionto calculate attendance
def attendance_cal(attendance):
  total=0
  for i in attendance:
    if i=="P":
      total=total+1
      #calculating percentage
    avg=total/len(attendance)*100
  print("Percentage:",avg)
  if avg>=75:
    return "Eligible"
  else:
    return "Not Eligible"

attendance=["P","P","A","A","P"]
result=attendance_cal(attendance)
print(result)
```

```python
"""6.Password Strength Checker:
Create a function to check password strength.
Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)"""
#functionto check the password
def password_checker(password):
  if len(password)<8:
    return "Weak password(Minimum 8 Characters required)"
  has_digit = False
  has_special = False

  # Loop through each character
  for char in password:
        if char.isdigit():
            has_digit = True

        if char in "@#$":
            has_special = True

  if has_digit and has_special:
      return "Strong Password"
  else:
      return "Weak Password (Must contain digit and special character)"
password = input("Enter password: ")
result = password_checker(password)
print(result)
```
